```python
df.rename(columns={'R&D Spend' : 'R&D_Spend','Marketing Spend':'Marketing_Spend'},inplace=True)
df
```

```python
X=df[['R&D_Spend','Administration','Marketing_Spend']]
y=df['Profit']
X,y
```

```python
pd.plotting.scatter_matrix(df,figsize=(10,10))
plt.show()
```

```python
from sklearn.linear_model import LinearRegression 
refressor=LinearRegression()
refressor
```

```python
from sklearn.model_selection import cross_val_score
nmse=cross_val_score(refressor,X,y,scoring='neg_mean_squared_error',cv=5)
nmse
np.mean(nmse)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge=Ridge()
parameters={'alpha':[1e-15,1e-10,1e-5,1e-2,1,5,10,20,30,40,50]}
ridge_regressor=GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error',cv=5)
```

```python
ridge_regressor.fit(X,y)
```

```python
ridge_regressor.best_score_
```
